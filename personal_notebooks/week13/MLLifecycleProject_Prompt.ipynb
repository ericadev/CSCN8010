### Prompt 1: Initial Context and Role Definition

Implementation notes:
- The prompt was developed using a technique known as Curriculum Learning. In the context of prompt engineering, this approach involves structuring prompts in a way that gradually increases in complexity. The user starts with simpler concepts and progressively introduces more complex ideas and terminology. This method helps the LLM build a solid foundation before tackling more advanced topics, ultimately enabling it to infer and generate new ideas based on its accumulated knowledge
